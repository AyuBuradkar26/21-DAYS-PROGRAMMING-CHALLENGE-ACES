#multiple plots in a graph

```python
fig=pt.figure()
rec=fig.patch
rec.set_facecolor("green")

x1=[1,2,3,4,5]
y1=[1,2,3,4,5]

x2=[1,2,3,4,5]
y2=[1,4,9,16,25]

x3=[1,2,3,4,5]
y3=[1,8,27,64,125]

graph1=fig.add_subplot(1,1,1)
graph1.plot(x1,y1,'red',linewidth=3.0)
graph1.plot(x2,y2,'orange',linewidth=3.0)
graph1.plot(x3,y3,'blue',linewidth=3.0)

pt.show()


```

```python
fig=pt.figure()
rec=fig.patch
rec.set_facecolor("green")

x1=[1,2,3,4,5]
y1=[1,2,3,4,5]

x2=[1,2,3,4,5]
y2=[1,4,9,16,25]

x3=[1,2,3,4,5]
y3=[1,8,27,64,125]

graph1=fig.add_subplot(3,1,1)
graph1.tick_params(axis="x",color="white")
graph1.tick_params(axis="y",color="white")

graph1.spines['top'].set_color('blue')
graph1.spines['left'].set_color('blue')
graph1.spines['bottom'].set_color('blue')
graph1.spines['right'].set_color('blue')

graph1.plot(x3,y3,'red',linewidth=3.0)

graph2=fig.add_subplot(3,1,2)#row,column,2nd plot
graph2.tick_params(axis="x",color="white")
graph2.tick_params(axis="y",color="white")

graph2.spines['top'].set_color('blue')
graph2.spines['left'].set_color('blue')
graph2.spines['bottom'].set_color('blue')
graph2.spines['right'].set_color('blue')

graph2.plot(x2,y2,'orange',linewidth=3.0)

graph3=fig.add_subplot(3,1,3)#row,column,2nd plot
graph3.tick_params(axis="x",color="white")
graph3.tick_params(axis="y",color="white")

graph3.spines['top'].set_color('blue')
graph3.spines['left'].set_color('blue')
graph3.spines['bottom'].set_color('blue')
graph3.spines['right'].set_color('blue')

graph3.plot(x1,y1,'purple',linewidth=3.0)

graph1.set_title("Y=X^3",color="yellow")
graph1.set_xlabel("X1-axis",color="yellow")
graph1.set_ylabel("Y1-axis",color="yellow")

graph2.set_title("Y=X^2",color="yellow")
graph2.set_xlabel("X2-axis",color="yellow")
graph2.set_ylabel("Y2-axis",color="yellow")


pt.show()
```

#bar graph

```python
import numpy as np
```

```python
po=np.arange(6)+0.5
names=['st1','st2','st3','st4','st5','st6']

pt.bar(po,(4,8,9,3,13,6),align='center',color='blue')

pt.xlabel("Student Names")
pt.ylabel("Height",color='red')
pt.title("Student Height details")

pt.tick_params(axis='x',color='red')
pt.tick_params(axis='y',color='red')

pt.xticks(po,names)
pt.show()
```

#Pi chart

```python
size=[50,23,15,7,5]#sum=100
labels=['android','apple','windows','xiomi','nokia']
colors=['yellow','orange','cyan','blue','red']

pt.pie(size,colors=colors,startangle=90,labels=labels)
pt.show()
```
